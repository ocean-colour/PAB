# PAB — getting started (Stage 0)

**PAB** produces matchup analyses between **PACE** (satellite ocean color) and **BGC-Argo** (autonomous float) data.

This notebook is the Stage 0 worked example: it shows the package imports, the `pab_version` provenance stamp, the configurable paths, and the package-version capture used in every fit record. It runs fully offline.

See the design (`docs/design/PAB_design.md`), coding plan (`docs/design/PAB_coding_plan.md`), and implementation record (`docs/design/PAB_implementation.md`).

## The package and its version

```python
import pab

# `pab_version` is the provenance stamp recorded with every fit
# (distinct from the packaging version in setup.py).
print('pab_version :', pab.pab_version)
print('__version__ :', pab.__version__)
```

## Configuration: paths and provenance

`pab.config` exposes the repo `ROOT` and a `DATA_DIR` (where the SQLite store and on-disk artifacts live; kept out of the repo by default and overridable via the `PAB_ROOT` / `PAB_DATA_DIR` environment variables).

```python
from pab.config import ROOT, DATA_DIR, package_versions

print('ROOT     :', ROOT)
print('DATA_DIR :', DATA_DIR)
```

`package_versions()` captures the versions of the scientific stack PAB depends on. These are stored per fit (Stage 5) so any result ties back to the exact code that produced it. Packages that are not installed are reported rather than raising.

```python
for name, ver in package_versions().items():
    print(f'{name:16s} {ver}')
```

## The subpackages

Stage 0 lays down the importable skeleton; each subpackage is filled in by its stage (db = Stage 1, argo = Stage 2, …).

```python
import importlib

for name in ['db', 'argo', 'pace', 'matchup', 'fit',
             'metrics', 'plotting', 'report']:
    mod = importlib.import_module(f'pab.{name}')
    print(f'pab.{name:9s} ->', (mod.__doc__ or '').strip().split(chr(10))[0])
```
